```python
import pandas as pd
```

```python
kospi = pd.read_csv("20201230 코스피 종목코드.csv", encoding="cp949")
kosdaq = pd.read_csv("20201230 코스닥 종목코드.csv",encoding="cp949")
iem_info = pd.read_csv('iem_info_20210902.csv')
```

```python
iem = iem_info.copy()
```

```python
iem
```

```python
iem.columns= ["종목코드","종목한글명","종목업종", "시가총액 규모유형","시장구분"]
```

```python
df1 = iem[iem['시장구분']==1]
df1['종목코드'].nunique()
```

```python
df2 = iem[iem['시장구분']==2]
```

```python
kospi.sort_values(by='종목코드' , ascending=True)
```

```python
iem_na = iem[iem['시장구분']==99]
iem_na.sort_values(by='종목코드' , ascending=True)
```

```python
'A' +str('%06d' % 20)
```

```python
kospi['종목코드'] = kospi['종목코드'].apply(lambda x: 'A'+str('%06d' % x))
```

```python
kospi.sort_values(by='종목코드' , ascending=True)
```

```python
df3 = pd.merge(kospi['종목코드'], iem_na, on='종목코드')
df3
```

```python
df3['시장구분']= df3['시장구분'].replace(99,3)
df3
```

```python
kosdaq['종목코드'] = kosdaq['종목코드'].apply(lambda x: 'A'+str('%06d' % x))
kosdaq
```

```python
df4 = pd.merge(kosdaq['종목코드'], iem_na, on='종목코드')
df4
```

```python
df4['시장구분']= df4['시장구분'].replace(99,4)
```

```python
df4
```

```python
iem_na_fill = pd.concat([df3,df4],ignore_index=True)
```

```python
iem_na_fill
```

```python
iem_na
```

```python
check = pd.merge(iem_na, iem_na_fill, on='종목코드', how='outer')
check
```

```python
iem_na_fill_na = check[check['종목업종_y'].isnull()]
iem_na_fill_na
```

```python
list(iem_na_fill_na['종목한글명_x'])
```

```python
iem_na_fill_na['종목한글명_x']
```

```python
iem_fill = pd.concat([df1, df2, iem_na_fill], ignore_index=True)
```

```python
iem_fill
```

```python
iem_mid = pd.merge( iem_fill , iem, how='outer', on='종목코드')
iem_mid
```

```python
iem_mid[iem_mid['시장구분_x'].isnull()].reset_index()
```

```python
iem_mid2 = iem_mid[iem_mid['시장구분_x'].isnull()].reset_index().drop(['index','종목한글명_x','종목업종_x','시가총액 규모유형_x','시장구분_x'], axis=1, inplace=False)
iem_mid2
```

```python
iem_mid2.columns=['종목코드','종목한글명','종목업종','시가총액 규모유형','시장구분']
```

```python
iem_result= pd.concat([iem_fill,iem_mid2],ignore_index=True)
iem_result
```

```python
iem_result['시장구분'].nunique()
```

```python
iem_result.to_csv("종목정보_시장정보처리.csv", encoding="cp949", index=False)
```
