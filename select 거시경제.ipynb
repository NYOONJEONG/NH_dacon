```python
oil_2015= pd.read_csv('원유가격20151201_20201231_달러.csv', encoding='cp949')
oil_2015['기간']= oil_2015['기간'].apply(lambda x: '20' + str(x[0:2]) + '-' + str(x[3:5]) + '-'+ str(x[6:8]))
oil_2015.rename(columns = {'기간' : 'Date'}, inplace = True)
oil_2015 = oil_2015.drop(oil_2015[oil_2015['WTI']=='-'].index)
oil_2015['WTI']= pd.to_numeric(oil_2015['WTI'])
oil_2015 = oil_2015.reset_index()
```

```python
import re

oil_14_8= pd.DataFrame(index=range(0,len(set(common['Date']))) ,columns =['Date']+['oil'])
n = 0

for i in tqdm(sorted(set(common['Date']))):
    temp_list = [np.nan]
    
    string = str(oil_2015[oil_2015['Date']== i].index)
    numbers = re.findall(r'\d+', string)
    temp = oil_2015.iloc[int(numbers[1])-14:int(numbers[1])]
    
    model = SimpleExpSmoothing(temp['WTI']).fit(smoothing_level=0.8, optimized=True ) # 단순지수평활 모델 생성
    fcast1 = model.forecast(1)
    temp_list.append(str(round(fcast1,2)).split()[1])
    
    oil_14_8.iloc[n,:]=temp_list
    n+=1
    
oil_14_8['Date'] = sorted(set(common['Date']))
```

```python
from sklearn.preprocessing import LabelEncoder
L_encoder = LabelEncoder()
L_encoder.fit(iem["종목한글명"]) # 종목 한글명
train_data["종목한글명"] = L_encoder.transform(train_data["종목한글명"])
test_data["종목한글명"] = L_encoder.transform(test_data["종목한글명"])
```

```python
train_data.reset_index(drop = True, inplace=True)
train_label.reset_index(drop = True, inplace=True)
```

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.model_selection import train_test_split, StratifiedKFold, KFold
from lightgbm import LGBMRegressor
```

```python
models = []

folds = KFold(n_splits=10)
for train_idx, val_idx in folds.split(train_data):
    
    train_x = train_data.iloc[train_idx, :]
    train_y = train_label[train_idx]
    val_x = train_data.iloc[val_idx, :]
    val_y = train_label[val_idx]
    
    model = LGBMRegressor(objective= "regression",
                          max_depth= 5,
                          n_estimators= 2000,
                          learning_rate= 0.01,
                          num_leaves = 31)
    
    model.fit(train_x, train_y,
              eval_set=[(val_x, val_y)],
              eval_metric=["rmse"],
              early_stopping_rounds=300,
              verbose=500)
    
    models.append(model)
```

```python
result = []
for i in models:
    result.append(i.predict(test_data))
    
predict = np.mean(result, axis = 0)
```
